Using the implemented function, calculate the entropy for each pair in this DataFrame and assign to the 'entropy' column.

```python
import numpy as np
import pandas as pd
 
 
def entropy(x):
    return np.round(-np.sum(x * np.log2(x)), 4)
 
df = pd.DataFrame(
    {
        'val_1': np.arange(0.01, 1.0, 0.1),
        'val_2': 1 - np.arange(0.01, 1.0, 0.1),
    }
)
df['entropy'] = [
    entropy([row[1][0], row[1][1]]) for row in df.iterrows()
]
print(df)
```
