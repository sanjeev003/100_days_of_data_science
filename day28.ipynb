1. np.char.replace(). This function is used to replace occurrences of a substring in a character array with another substring.

import numpy as np
A = np.array(['apple', 'banana', 'orange', 'pineapple'])
replaced = np.char.replace(A, 'a', 'x')
print(replaced)

> Output :- 
    ['xpple' 'bxnxnx' 'orxnge' 'pinexpple']

2. The np.savetxt() function in NumPy is used to save an array to a text file. It saves the array as plain text, with each row of the array corresponding to a line in the file.

    np.savetxt(fname, X, fmt='%.18e', delimiter=' ', newline='\n', header='', footer='', comments='# ')
    fname: The file name or file object where the array is to be saved.
    X: The array to be saved.
    fmt: The format of the array elements. It is a string that specifies how the elements should be formatted when written to the file.
    delimiter: The string used to separate the elements in each row of the array.
    newline: The string used to separate the lines in the file.
    header: The string to be written at the beginning of the file.
    footer: The string to be written at the end of the file.
    comments: The string used to indicate comments in the file.

import numpy as np
A = np.array([[1, 2, 3],
              [4, 5, 6]])
np.savetxt('data.txt', A, fmt='%d', delimiter=',')

> Output :- 
    1,2,3
    4,5,6

3. The np.loadtxt() function in NumPy is used to load data from a text file into a NumPy array.

    np.loadtxt(fname, dtype=<class 'float'>, comments='#', delimiter=None, converters=None, skiprows=0, usecols=None, unpack=False, ndmin=0)
    fname: The file name or file object from which the data will be loaded.
    dtype: The data type of the resulting array. By default, it is set to float.
    comments: The character or characters used to indicate comments in the file. Lines starting with the specified character(s) will be ignored.
    delimiter: The string used to separate the values in each row of the file. By default, any whitespace is used as the delimiter.
    converters: A dictionary mapping column indices or names to conversion functions. These functions are applied to the corresponding column(s) in the file.
    skiprows: The number of rows to skip from the beginning of the file.
    usecols: A sequence specifying which columns to extract from the file. If None, all columns are read.
    unpack: If True, the returned array is transposed. This is useful when loading data with multiple columns.

import numpy as np
data = np.loadtxt('data.txt', delimiter=',')
print(data)

> Output :- 
    array([[1., 2., 3.],
       [4., 5., 6.]])

```python
import numpy as np

np.set_printoptions(linewidth=160, suppress=True)

wig_games_raw = """Profile	Time	Course	Change	Change%	Ref.	Open	Low	High	Volume	Turn	Share
11B (11BIT)	17 kwi 17:01	391.00	+8.00	(+2.09%)	383.00	383.50	383.00	394.00	12 784	4 994 874	19.034%
CDR (CDPROJEKT)	17 kwi 17:01	339.50	+5.30	(+1.59%)	334.20	338.30	337.00	344.20	233 059	79 245 368	39.618%
CIG (CIGAMES)	17 kwi 17:03	0.742	-0.012	(-1.59%)	0.754	0.772	0.730	0.772	1 311 078	971 459	1.855%
PLW (PLAYWAY)	17 kwi 17:03	387.50	+18.00	(+4.87%)	369.50	374.00	373.00	388.00	33 206	12 661 786	10.638%
TEN (TSGAMES)	17 kwi 17:02	349.50	+22.50	(+6.88%)	327.00	332.00	330.00	353.50	39 793	13 697 060	28.855%"""

lines = wig_games_raw.splitlines()
lines = [line.split('\t') for line in lines]
wig_games = np.array(lines, dtype=np.str)
wig_games = np.char.replace(wig_games, ' ', '')
print(wig_games)
```
